# A company started to invest in digital marketing as a new way of their product promotions.For that they collected data and decided to carry out a study on it.

# ● The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

```python
from scipy import stats
```

```python
# H0 : There is no increase in sales after stepping into digital marketing
# H1 : There is increase in sales after stepping into digital market
```

```python
t_test,pval = stats.ttest_ind(data['Sales_before_digital_add(in $)'],data['Sales_After_digital_add(in $)'])
print(t_test,pval)

if(pval>0.05):
    print("we accept the null hypothesis that the Sales haven't changed after stepping into digital marketing")
else:
    print("we reject the null hypothesis and accept the alternate hypothesis,that there is increase in sales after stepping into digital marketing")
```

# ● The company needs to check whether there is any dependency between the features “Region” and “Manager”.


```python

f1 = data['Region']
f2 = data['Manager']
```

```python
# H0 : The feature 'Region' and 'Manager' are independent 
# H1 : The feature 'Region' and 'Manager' are dependent
```

```python
# Creating crosstable
m_Crosstable = pd.crosstab(f1,f2)
m_Crosstable
```

```python
from scipy.stats import chi2_contingency
```

```python
chiVal, pVal, df,exp = chi2_contingency(m_Crosstable)
chiVal,pVal, df,exp
```

```python
# So we accept the null hypothesis, ie  The feature 'Region' and 'Manager' are independent
```
